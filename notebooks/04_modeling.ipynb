```python
import sys
import pandas as pd

# Trỏ đường dẫn về thư mục gốc
sys.path.append('..')

from src.data.loader import fetch_and_load_data
from src.data.cleaner import clean_and_preprocess
from src.models.supervised import run_classification_models, run_regression_timeseries

# 1. Nạp dữ liệu
df_raw = fetch_and_load_data()
df_processed = clean_and_preprocess(df_raw)

# 2. Chạy bài toán Phân Lớp (Dự đoán lỗi)
print("="*50)
print("BÀI TOÁN PHÂN LỚP (DỰ ĐOÁN MACHINE FAILURE)")
print("="*50)
class_metrics, error_analysis = run_classification_models(df_processed, df_raw)

print("1. Đánh giá mô hình (PR-AUC & F1-Score):")
for model, scores in class_metrics.items():
    print(f"- {model}: PR-AUC = {scores['PR-AUC']:.3f} | F1-Score = {scores['F1-Score']:.3f}")

print("\n2. Phân tích các loại lỗi bị bỏ sót (False Negatives) của Random Forest:")
for error_type, count in error_analysis.items():
    print(f"- Bỏ sót {error_type}: {count} ca")

# 3. Chạy bài toán Hồi Quy / Chuỗi thời gian (Dự đoán mòn dụng cụ)
print("\n" + "="*50)
print("BÀI TOÁN HỒI QUY CHUỖI THỜI GIAN (DỰ ĐOÁN TOOL WEAR)")
print("="*50)
reg_metrics = run_regression_timeseries(df_processed, df_raw)

print("Đánh giá mô hình Random Forest Regressor (Không Shuffle, có Lag-features):")
print(f"- MAE (Sai số tuyệt đối trung bình): {reg_metrics['MAE']:.2f} phút")
print(f"- RMSE (Sai số bình phương trung bình gốc): {reg_metrics['RMSE']:.2f} phút")
```
