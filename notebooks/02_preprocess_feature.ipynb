```python
# Load dữ liệu và Tiền xử lý cơ bản
import sys
import pandas as pd
import os

# Thêm đường dẫn để nhận diện thư mục src
sys.path.append('..')
from src.data.loader import fetch_and_load_data
from src.data.cleaner import clean_and_preprocess

# 1. Tải dữ liệu gốc
df_raw = fetch_and_load_data()

# 2. Tiền xử lý (Xóa leakage, Encode, Scaling)
df_clean = clean_and_preprocess(df_raw)
print("Dữ liệu sau khi làm sạch cơ bản:", df_clean.shape)
```

```python
# Tạo đặc trưng trễ (Lag Features) cho bài toán Hồi quy
df_features = df_clean.copy()
df_features['UDI'] = df_raw['UDI']
df_features = df_features.sort_values(by='UDI')

# Tạo độ trễ 1 chu kỳ cho các thông số cảm biến
cols_to_lag = ['Air temperature [K]', 'Rotational speed [rpm]', 'Torque [Nm]']
for col in cols_to_lag:
    df_features[f'{col}_lag1'] = df_features[col].shift(1)

df_features = df_features.dropna()

# Lưu dữ liệu "vàng" này lại để các bước Mining/Modeling sử dụng đồng nhất
os.makedirs('../data/processed/', exist_ok=True)
df_features.to_csv('../data/processed/final_features.csv', index=False)
print("Đã lưu dữ liệu đặc trưng tại: data/processed/final_features.csv")
df_features.head()
```
